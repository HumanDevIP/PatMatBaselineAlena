```python
import pandas as pd
import json
```

```python
# open dataframe
df = pd.read_excel('../data/tweet_sentiment_input_file.xlsx')
```

```python
df.head()
```

```python
# select rows with extreme sentiment values
filtered_df = df[df['sentiment'].isin([0, 3])]
```

```python
# count rows with extreme values
sentiment_0 = df['sentiment'].value_counts().get(0, 0)
sentiment_3 = df['sentiment'].value_counts().get(3, 0)
print(sentiment_0, sentiment_3)
```

```python
# randomly select 187 lines with sentiment 3
sentiment_3_data = df[df['sentiment'] == 3].sample(n=187, random_state=42)
sentiment_0_data = df[df['sentiment'] == 0]
```

```python
df_data = pd.concat([sentiment_3_data, sentiment_0_data])
```

```python
df_data.loc[df_data['sentiment'] == 3, 'sentiment'] = 1
```

```python
labels = df_data['sentiment'].tolist()
texts = df_data['text'].tolist()
```

```python
with open('../data/labels.json', 'w') as lb:
  json.dump(labels, lb)
```

```python
with open('../data/texts.json', 'w') as tx:
  json.dump(texts, tx)
```
